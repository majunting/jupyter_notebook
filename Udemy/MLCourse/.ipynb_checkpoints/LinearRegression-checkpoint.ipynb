# Linear Regression

Let's fabricate some data that shows a roughly linear relationship between page speed and amount purchased:

```python
%matplotlib inline
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

scatter(pageSpeeds, purchaseAmount)
```

As we only have two features, we can keep it simple and just use scipy.state.linregress:

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

Not surprisngly, our R-squared value shows a really good fit:

```python
r_value ** 2
```

Let's use the slope and intercept we got from the regression to plot predicted values vs. observed:

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```

## Activity

Try increasing the random variation in the test data, and see what effect it has on the r-squared error value.

```python
new_pageSpeeds = np.random.normal(3.0, 0.01, 1000)
new_purchaseAmount = 100 - (new_pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3
# new_pageSpeeds = np.random.normal(3.0, 10, 1000)
# new_purchaseAmount = 100 - (new_pageSpeeds + np.random.normal(0, 10, 1000)) * 3

scatter(new_pageSpeeds, new_purchaseAmount)

slope, intercept, r_value, p_value, std_err = stats.linregress(new_pageSpeeds, new_purchaseAmount)

print(r_value ** 2)

fitLine = predict(new_pageSpeeds)

plt.scatter(new_pageSpeeds, new_purchaseAmount)
plt.plot(new_pageSpeeds, fitLine, c='r')
plt.show()
```
